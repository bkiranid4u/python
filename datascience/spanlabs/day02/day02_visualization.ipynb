```python
import pandas as pd 
import matplotlib as mpl
```

```python
dataset_url = 'https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv'
sdf = pd.read_csv(dataset_url,parse_dates=['Date'],index_col='Date')
sdf.head(5)
```

```python
sdf.index
```

```python
sdf.plot(y='MSFT', figsize=(10,6), title='Microsoft Stock', ylabel='USD')
```

```python
sdf.plot.line(y=['FB','AAPL','MSFT'], figsize=(10,6))
```

```python
sdf_3m = sdf.resample(rule='M').mean()[-3:]
```

```python
sdf_3m.plot( figsize=(10,6), title='Microsoft Stock', ylabel='USD')
```

```python
sdf_3m.plot.bar(figsize=(10,6), ylabel='Price', color=['red','green','blue'])
```

```python
sdf_3m.plot.barh(figsize=(10,6), ylabel='Price', color=['red','green','blue'])
```

```python
sdf_3m.plot(kind='bar', stacked=True, figsize=(9,6),color=['red','green','blue'])
```

```python
sdf[['MSFT', 'FB']].plot(kind='hist', bins=25, figsize=(9,6))
```

```python
sdf.plot(y=['MSFT','FB'],kind='hist', bins=25, figsize=(9,6))
```

```python
sdf.plot.box(figsize=(9,6))
```

```python
sdf.plot(kind='area', figsize=(9,6))
```

```python
sdf_3m.index = ['March', 'April', 'May']
sdf_3m.plot(kind='pie',y='AAPL', legend=True, autopct='%.f')
```
