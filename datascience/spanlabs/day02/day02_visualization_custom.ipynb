```python
import pandas as pd
import matplotlib
```

```python
dataset_url = 'https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv'
sdf = pd.read_csv(dataset_url, parse_dates=['Date'], index_col='Date')
sdf.head(5)
```

```python
def format_axis(label):
    result = label.month_name() + "-" + str(label.year)
    return result
    
```

```python
sdf_3m = sdf.resample(rule='M').mean()[-3:]
```

```python
ax = sdf_3m.plot(kind='bar', figsize=(10,6), ylabel='Price')
ax.set_xticklabels(map(format_axis,sdf_3m.index))
```
