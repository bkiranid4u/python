```python
import pandas  as pd 
```

```python
bank = pd.read_csv('/Users/kiran/Documents/GitHub/python/datascience/spanlabs/data/bank/bank.csv', sep=';')
bank.dtypes
```

```python
bank.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in bank.select_dtypes(include='object'):
    bank[column] = le.fit_transform(bank[column])
    
bank.head()
```

```python
x = bank.drop('y', axis=1)
y = bank['y']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=16)
```

```python
from sklearn.linear_model import LogisticRegression
# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
```

```python
from sklearn.metrics import classification_report
target_names = ['Default', 'No Default']
print(classification_report(y_test, y_pred, target_names=target_names))
```
