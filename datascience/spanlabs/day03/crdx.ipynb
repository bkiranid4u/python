```python
import pandas as pd 
```

```python
crx = pd.read_csv(
    '/Users/kiran/Documents/GitHub/python/datascience/spanlabs/data/credit/crx.data', header=0)

crx.columns = ['A1',
'A2',
'A3',
'A4',
'A5',
'A6',
'A7',
'A8',
'A9',
'A10',
'A11',
'A12',
'A13',
'A14',
'A15',
'A16']
```

```python
columns = ['A1','A2','A4','A5','A6','A7','A14']

for column in columns:
    crx.drop(crx[crx[column] == '?'].index, inplace=True)
crx.head()
```

```python
crx.dtypes

```

```python
crx.A2 = crx.A2.astype('float')
crx.A14 = crx.A14.astype('int')
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in crx.select_dtypes(include='object'):
    print(column)
    crx[column] = le.fit_transform(crx[column])

crx.head()
```

```python
from sklearn.model_selection import train_test_split
x = crx.drop('A16', axis=1)
y = crx['A16']


x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)

y_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_test, y_pred))
```
