```python
hep = pd.read_csv('/Users/kiran/Documents/GitHub/python/datascience/spanlabs/data/hepatitis/hepatitis.data', header=0)
hep.columns = ['Class',
               'AGE',
               'SEX',
               'STEROID',
               'ANTIVIRALS',
               'FATIGUE',
               'MALAISE',
               'ANOREXIA',
               'LIVER BIG',
               'LIVER FIRM',
               'SPLEEN PALPABLE',
               'SPIDERS',
               'ASCITES',
               'VARICES',
               'BILIRUBIN',
               'ALK PHOSPHATE',
               'SGOT',
               'ALBUMIN',
               'PROTIME',
               'HISTOLOGY',]
```

```python

for column in hep.columns:
    hep.drop(hep[hep[column] == '?'].index, inplace=True)
hep.head()

hep.BILIRUBIN = hep.BILIRUBIN.astype('float')
hep.ALBUMIN = hep.ALBUMIN.astype('float')
```

```python
hep.BILIRUBIN = hep.BILIRUBIN.astype('float')
hep.ALBUMIN = hep.ALBUMIN.astype('float')
```

```python
for column in hep.select_dtypes(include='object'):
    hep[column] = hep[column].astype('int64')

hep.dtypes
```

```python
from sklearn.model_selection import train_test_split
x = hep.drop('Class', axis=1)
y = hep['Class']


x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)

y_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_test, y_pred))
```
