```python
import pandas as pd 
```

```python
bank = pd.read_csv('https://raw.githubusercontent.com/vishurudratrainer/notebooks/main/bank.csv',sep=';')
```

```python
bank.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact',
        'month', 'poutcome', 'y']

le = LabelEncoder()

for col in cols:
    
    bank[col] = le.fit_transform(bank[col])

```

```python
from sklearn.model_selection import train_test_split

x = bank.drop('y',axis=1)

y = bank['y']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=0 )
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
auto = pd.read_csv(
    '/Users/kiran/Documents/GitHub/python/datascience/spanlabs/data/auto/imports-85.data',header=None)
```

```python
auto.columns = ['symboling',
                'normalized_losses',
                'make',
                'fuel_type',
                'aspiration',
                'num_of_doors',
                'body_style',
                'drive_wheels',
                'engine_location',
                'wheel_base',
                'length',
                'width',
                'height',
                'curb_weight',
                'engine_type',
                'num_of_cylinders',
                'engine_size',
                'fuel_system',
                'bore',
                'stroke',
                'compression_ratio',
                'horsepower',
                'peak_rpm',
                'city_mpg',
                'highway_mpg',
                'price']
```

```python
auto.head(10)
```

```python
auto.drop(auto[auto['normalized_losses'] == "?"].index, inplace=True)
auto.drop(auto[auto['bore'] == "?"].index, inplace=True)

auto.drop(auto[auto['bore'] == "?"].index, inplace=True)

auto['normalized_losses'] = auto['normalized_losses'].astype('int')
auto.bore = auto.bore.astype('float')
auto.stroke = auto.stroke.astype('float')
auto.horsepower = auto.horsepower.astype('int')
auto.peak_rpm = auto.peak_rpm.astype('int')
auto.price = auto.price.astype('int')
```

```python
auto.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for column in auto.select_dtypes(include='object').columns:
    print(column)
    auto[column] = le.fit_transform(auto[column])
```

```python
x = auto.drop('price',axis=1)
y = auto['price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(x_test)

print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
