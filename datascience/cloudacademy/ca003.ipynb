```python
import pandas as pd
import sqlite3

db = sqlite3.connect('data/Northwind_large.sqlite')
cur = db.cursor()

df = pd.read_sql_query('''
    SELECT * 
    FROM Customer
''',db)

df.head()
```

```python
df = pd.read_sql_query('''
    SELECT [ContactName], [Address]
    FROM [Customer]
    where [City] = 'London';
''',db)
df.head()
```

```python
df = pd.read_sql_query('''
    SELECT 
        o.[Id] AS [OrderId],
        c.[ContactName],
        c.[Phone],
        o.[OrderDate],
        o.[ShippedDate]
    FROM
        [Order] AS o
        JOIN
        [Customer] AS c
        ON
        o.CustomerId = c.Id 
    WHERE 
        date(o.ShippedDate) > date('2015-07-01')
''',db)
df.head()
```
